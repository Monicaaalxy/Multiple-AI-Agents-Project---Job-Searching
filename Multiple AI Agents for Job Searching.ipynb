# Agents

### Job Researcher

```python
researcher = Agent(
   role="Data Science/Quantitative Analyst/Machine Learning Job Research",
   goal="You should conduct in-depth analysis of posted jobs to help job candidates",
   backstory=("You're an excellent job research. You have competancy in discovering and extracting critical information from job posts"
              "You help job candidates pinpoint necessary qualifications and skills chased by recruiters."
              "This will play an important role in helping job candidates design their resumes."),
    tools=[scrape_tool,search_tool],
    verbose=True
)
```

### Candidate profile research

```python
profile_researcher = Agent(
  role="Personal Profiler for Data Scientists/Quantitative Analysts/Machine Learning Engineers",
  goal="Performan in-depth research on job applicants based on their resumes. Help them stand out the job market",
  backstory=("Equipped with incredible analytical ability, you collect and synthesize information from a variety of sources to "
             "craft comprehensive personal and professional profiles."
             "This will lay the foundation for customized resume improvement"),
 tools=[scrape_tool,search_tool,read_resume,semantic_search_resume],
 verbose=True
)
```

### Resume customization

```python
resume_tailor = Agent(
 role="Resume Tailors for Data Scientists/Quantitative Analysts/Machine Learning Engineers",
 goal="Try all your best to make a resume stand out in the job market",
 backstory=("You're a perfect tailor to customize job candidates' resumes. You have a sharp and strategic eye for both big picture and details"
            "You excel at refining candidates' resumes by highlighting the most relevant skills and experiences"
            "You can make sure a very high compatibility between candidates and jobs they applied."),
 tools=[scrape_tool,search_tool,read_resume,semantic_search_resume],
 verbose=True
)
```

# Tasks

### Job Research

```python
job_research=Task(
 description=("1. Analyze the job post here {post}"
              "2. Based on the job post, extract key skills, qualifications, experiences."
              "3. Use the given tool to collect content and categorize these requirement."),
 expected_output= ("A structured list of job requirements that include necessary skills, qualifications, and experiences"),
 agent=researcher,
 async_execution=True
)
```

```python
profiling=Task(
    description=("1. Compile a detained and professional profile by referring to my resume: {link} and introduction: {intro}."
                 "2. Utilize tools to extract and synthesize information from these sources"),
    
 expected_output= ( "A comprehensive profile document that includes skills, "
        "project experiences, contributions, interests, and "
        "communication style."),
 agent=profile_researcher,
 async_execution=True   
)
```

```python
resume_customization=Task(
 description=("1. Using the profile and job requirements obtained from "
              "previous tasks, tailor the resume to highlight the most "
              "relevant areas."
              "2. Employ tools to adjust and enhance the "
              "resume content. Make sure this is the most compatible resume to the job posting but "
              "don't make up any information."
              "3. Update every section on my resume to make sure the content is compatible to the job posting."),
 expected_output="An updated resume that effectively highlights the candidate's "
                 "qualifications and experiences relevant to the job.",
 agent=resume_tailor,
 output_file="Resume-Monica.md",
 context=[job_research,profiling]
)
```

# Crew

```python
crew = Crew(
    agents=[researcher,profile_researcher,resume_tailor],
    
    tasks=[job_research, profiling,resume_customization],
    verbose=True
)
```
